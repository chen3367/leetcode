# Non Overlapping Intervals (TSMC Hackerrank)
***
Three new health care buildings are to be built in HackerLand. They will be built on land with plots numbered consecutively starting from 1. There are two integer arrays $starting$, $ending$, each of size $n$, that represent= $n$ intervals where the $i^{th}$ intervals is [starting[i], ending[i]]. Determine the number of ways three non-overlapping intervals can be selected from the $n$ intervals.

### **Note**:
- An Interval is defined as $[l, r]$ where it contains every integer between $l$ to $r$.
- Two intervals $[l[1], r[1]]$ and $[l[2], r[2]]$ overlap if there is an integer that is contained in both of them. For example, [2, 5], [3, 8] overlap because 3, 4, 5 are common to both, whereas [1, 5], [8, 9] do non-overlapping.
- Interval sets must be distinct. At least on interval must be different for two sets to be distinct. For example, interval sets [[1, 2], [3, 4]] and [[3, 4], [1, 2]] share the same intervals in a different order, so they are not distinct.

### **Example**
$n = 5$ <br>
$starting = [1, 2, 4, 3, 7]$ <br>
$ending = [3, 4, 6, 5, 8]$ <br>

<br>
Expected output: 1

There are 5 intervals [[1, 3], [2, 4], [4, 6], [3, 5], [7, 8]]. They are shown as line segments 1 through 5.

There is only one way to select three non-overlapping intervals, lines 1, 3, and 5 with ranges [[1, 3], [4, 6], [7, 8]].<br>
Note that we cannot select line 1, 4 and 5 with ranges [[1, 3], [3, 5], [7, 8]] becausse [1, 3] and [3, 5] overlap at 3.

### **Function Description**
Complete the function *getThreeNonOverlappingIntervals* in the editor below.

*getThreeNonOverlappingIntervals* has the following parameter(s): <br>
$\;\;\;\;\;\;$ *int starting[n]*: the left endpoints of the intervals <br>
$\;\;\;\;\;\;$ *int ending[n]*: the right endpoints of the intervals

### **Returns**
$\;\;\;\;\;\;$ *long* the number of ways of selecting three non-overlapping intervals

### **Constraints**
- $1 \le n \le 10^5$
- $|starting|, |ending| = n$
- $1 \le starting[i], ending[i] \le 10^9$

### > **Sample Case 0**
### **Sample Input 0**
$starting = [5, 2, 3, 7]$<br>
$ending = [7, 2, 4, 8]$<br>
### **Sample Output 0**
2<br>
<br>
There are 2 ways to select three non-overlapping intervals.<br>
[5, 7], [2, 2], [3, 4]<br>
[2, 2], [3, 4], [7, 8]<br>

```python
starting = [5, 2, 3, 7]
ending = [7, 2, 4, 8]
```

```python
def getThreeNonOverlappingIntervals(starting, ending):
    intervals = sorted([[s, e] for s, e in zip(starting, ending)])
    n = len(intervals)
    count = 0
    for i in range(n):
        for j in range(i, n):
            if intervals[i][1] >= intervals[j][0]:
                continue
            for k in range(j, n):
                if intervals[j][1] < intervals[k][0]:
                    count += len(intervals[k:])
                    break
    return count
```

```python
print(getThreeNonOverlappingIntervals(starting, ending))
```
